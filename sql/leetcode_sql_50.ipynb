#### 1757. Recyclable and Low Fat Products

Write a solution to find the ids of products that are both low fat and recyclable

```python
SELECT 
    product_id
FROM Products
WHERE low_fats = 'Y' AND recyclable = 'Y'
```

#### 584. Find Customer Referee

Find the names of the customer that are either:

1. referred by any customer with id != 2.
2. not referred by any customer.

```python
SELECT 
    name
FROM Customer
WHERE referee_id != 2 OR referee_id IS NULL
```

#### 595. Big Countries

A country is big if:

1. it has an area of at least three million (i.e., 3000000 km2), or
2. it has a population of at least twenty-five million (i.e., 25000000).

```python
SELECT 
    name
    , population
    , area
FROM World
WHERE area >= 3000000 OR population >= 25000000
```

#### 1148. Article Views I

Write a solution to find all the authors that viewed at least one of their own articles.

```python
SELECT
    distinct author_id as id
FROM Views
WHERE viewer_id = author_id
ORDER BY author_id ASC
```

#### 1683. Invalid Tweets

Write a solution to find the IDs of the invalid tweets. The tweet is invalid if the number of characters used in the content of the tweet is strictly greater than 15.

```python
SELECT
    tweet_id
FROM Tweets
WHERE LENGTH(content) > 15
```

#### 1378. Replace Employee ID With The Unique Identifier

Write a solution to show the unique ID of each user, If a user does not have a unique ID replace just show null.

```python
SELECT
    eu.unique_id
    , e.name
FROM Employees AS e
LEFT JOIN EmployeeUNI eu ON eu.id = e.id
```

#### 1068. Product Sales Analysis I

Write a solution to report the product_name, year, and price for each sale_id in the Sales table.

```python
SELECT 
    p.product_name
    , s.year
    , s.price
FROM Sales AS s
JOIN Product AS p on p.product_id = s.product_id
```

#### 1581. Customer Who Visited but Did Not Make Any Transactions

Write a solution to find the IDs of the users who visited without making any transactions and the number of times they made these types of visits.

```python
SELECT
    v.customer_id
    , COUNT(*) AS count_no_trans
FROM Visits AS v
LEFT JOIN Transactions AS t ON t.visit_id = v.visit_id
WHERE t.transaction_id IS NULL
GROUP BY v.customer_id
```

#### 197. Rising Temperature

Write a solution to find all dates' id with higher temperatures compared to its previous dates (yesterday).

```python
SELECT
    today.id
FROM Weather AS today
JOIN Weather AS yesterday ON yesterday.recordDate = today.recordDate - INTERVAL '1 day'
WHERE today.temperature > yesterday.temperature
```
