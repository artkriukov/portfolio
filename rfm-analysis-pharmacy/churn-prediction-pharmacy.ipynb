5. Feature Engineering

Расчёт:

- №	Признак	Что сделать с ним перед обучением
- 1	recency_days	Стандартизация (StandardScaler)
- 2	purchase_cnt_90d	Стандартизация (или логарифмирование, если сильный перекос)
- 3	revenue_90d	Логарифм (log1p), так как выручка всегда скошена вправо
- 4	avg_check	Стандартизация
- 5	avg_days_between_purchases	Заменить бесконечности (если была 1 покупка) на max_gap + 1, затем стандартизация
- 6	bonus_usage_ratio	КРИТИЧНО: Если bonus_earned = 0, то ratio должен быть равен -1 (или 0, но явно обработайте NaN)
- 7	unique_products	Стандартизация
- 8	activity_decline	Стандартизация. Проверьте, чтобы не было деления на ноль (если 90d=0) — поставьте 0
- 9	revenue_decline	Стандартизация. Аналогично, обработайте нули
- 10	max_gap	Если был всего 1 визит, поставьте recency_days (текущий перерыв)
- 11	tenure	Стандартизация
- 12	rfm_segment	One-Hot Encoding (превратить в 4-5 бинарных столбцов) или Label Encoding, если это упорядоченные сегменты (Champion=5, Loyal=4, At Risk=2 и т.д.)
